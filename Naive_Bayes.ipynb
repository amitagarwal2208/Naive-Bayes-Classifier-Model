## Building our Naive Bayes Classifier

```python
def priorProb(ytrain , label):
    
    totalEx = ytrain.shape[0]
    classEx = np.sum(ytrain==label)
    
    return (classEx/totalEx)


def cond_prob(xtrain , ytrain , featureCol , featureVal , label):
    
    
    mylabel_x = xtrain[ytrain == label]
    numerator = np.sum(mylabel_x[:,featureCol]==featureVal)
    denominator = mylabel_x.shape[0]
    
    return numerator/float(denominator)

## Computing Posterior probability for each test example and make predictions.


def predict(xtrain , ytrain , xtest):
    # xtest : a single testing point , n features.
    
    classes = np.unique(ytrain)
    numOfFeatures = xtrain.shape[1]
    postProbs = []
    
    for label in classes:
        
        prior = priorProb(ytrain , label)
        likelihood = 1
        for f in range(numOfFeatures):
            likelihood*=cond_prob(xtrain,ytrain,f,xtest[f],label)
    
        
        postProbs.append(prior*likelihood)
        
    pred = np.argmax(postProbs)
    return pred 
    
    
    
def accuracyScore(xtrain , ytrain , xtest , ytest):
    
    ypred=[]
    
    for i in range(xtest.shape[0]):
        predLabel = predict(xtrain , ytrain , xtest[i])
        ypred.append(predLabel)
        
    accuracy = np.sum(ytest==ypred)/ytest.shape[0]
    return accuracy

    

    
    
    
    
```

```python
print(accuracyScore(xtrain , ytrain , xtest , ytest))
```
